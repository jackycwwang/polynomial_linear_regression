### Polynomial Regression

If we have a polynomial, let's say a quadratic or cubic, or any other type of curved functions, we can still use linear regression, despite its name. Linear regression means that the weight parameters are linear. it doesn't say anything about X has to be. If you look at let's say quadratic function or ax^2 + bx + c, you see that y is linear in a, b, c. This is the idea of how we might apply linear regression to a polynomial. This is also why it's important to plot your data before trying to fit a model.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
X = []
Y = []
for line in open('data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x])    # we need a bias term and also the x^2 term
    Y.append(float(y))
```

```python
# convert to numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Plot the original data to see what it looks like
# It is a quadratic function as we expect
plt.scatter(X[:,1], Y);
```

```python
# Calculate weights
# Notice this is also multiple linear regression so we can do the exact 
# thing as we did before. The only difference is how we created the input table X
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
```

```python
Yhat = X.dot(w)
```

```python
# note X is not sorted, it draws lines between each of the points
plt.scatter(X[:, 1], Y)
plt.plot(X[:, 1], Yhat);
```

```python
# working now because quadratic function is monotonically increasing
plt.scatter(X[:, 1], Y)
plt.plot(sorted(X[:, 1]), sorted(Yhat), c='r');
```

```python
# Calculate R-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1)/d2.dot(d2)
```
